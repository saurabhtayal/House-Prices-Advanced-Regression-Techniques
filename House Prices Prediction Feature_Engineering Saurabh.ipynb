## Feature Engineering
We will be performing all the below steps in Feature Engineering:

- Missing values
- Temporal variables
- Categorical variables: remove rare labels
- Standarise the values of the variables to the same range

###### We should always do the feature engineering on train and test dataset separately so that no information from test data is seen by the system along with the train data before modelling.

### Missing Values

```python
# Let us capture all the nan values
# First lets handle Categorical features which are missing
print("In Train Data:-\n% Missing values in Categorial Variables:-\n")
features_nan = [feature for feature in train.columns if train[feature].isnull().sum() > 0 and train[feature].dtypes == 'O']

for feature in features_nan:
    print("{}: {}% missing values".format(feature, np.round(train[feature].isnull().sum() / train.shape[0] * 100, 3)))
```

```python
# Replace missing value with a new label both in train and test datasets
for dataset in combined:
    features_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 0 and dataset[feature].dtypes == 'O']
    dataset[features_nan] = dataset[features_nan].fillna('Missing')
    
    # checking the null values after filling them both in train and test data
    print(dataset[features_nan].isnull().sum(), end = '\n\n')
```

```python
# Now lets check for numerical variables the contains missing values
print("In Train Data:-\n% Missing values in Numerical Variables:-\n")
numerical_with_nan_train = [feature for feature in train.columns if train[feature].isnull().sum() > 0 and train[feature].dtypes != 'O']

for feature in numerical_with_nan_train:
    print("{}: {}% missing values".format(feature, np.round(train[feature].isnull().sum() / train.shape[0] * 100, 3)))
```

```python
# Now lets check for numerical variables the contains missing values
print("In Test Data:-\n% Missing values in Numerical Variables:-\n")
numerical_with_nan_test = [feature for feature in test.columns if test[feature].isnull().sum() > 0 and test[feature].dtypes != 'O']

for feature in numerical_with_nan_test:
    print("{}: {}% missing values".format(feature, np.round(test[feature].isnull().sum() / test.shape[0] * 100, 3)))
```

```python
numerical_with_nan = numerical_with_nan_train + numerical_with_nan_test
```

```python
# Replacing the numerical Missing Values in both train and test data
for dataset in combined:
    for feature in numerical_with_nan:
        ## We will replace by using median since there are outliers
        median_value = dataset[feature].median()
    
        ## create a new feature to capture nan values
        dataset[feature + 'nan'] = np.where(dataset[feature].isnull(),1 , 0)
        dataset[feature].fillna(median_value, inplace = True)
    
    print(dataset[numerical_with_nan].isnull().sum(), end = '\n\n')
```

```python
print(train.isna().sum().sum(), test.isna().sum().sum())
```

### Temporal Variables (Date-Time Variables)

```python
for dataset in combined:
    for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
        dataset[feature] = dataset['YrSold'] - dataset[feature]
```

### Numerical Variables

#### Since the numerical features are skewed, we will perform log transformation

```python
display(train.head())
```

```python
# Train Data
num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']
for feature in num_features:
    train[feature] = np.log(train[feature])
# Test Data
num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea']
for feature in num_features:
    test[feature] = np.log(test[feature])
```

```python
print(train.isna().sum().sum(), test.isna().sum().sum())
```

### Handling Rare Categorical Features

#### We will replace those categories from the categorical variables which are less than 1% of the observations with a new common name 'Rare'

```python
categorical_features = [feature for feature in train.columns if train[feature].dtype == 'O']
print(categorical_features)
```

```python
for dataset in combined:
    for feature in categorical_features:
        temp = dataset.groupby(feature)['Id'].count() / len(dataset)
        temp_df = temp[temp > 0.01].index
        dataset[feature] = np.where(dataset[feature].isin(temp_df), dataset[feature], 'Rare_val')
```

#### Converting Categorical Variables

```python
df = pd.concat([train, test], sort = False)
print(df.shape)
display(df.head())
```

```python
df_dum = pd.get_dummies(df, drop_first = True)
print(df_dum.shape)
display(df_dum.tail())
```

```python
train_dum = df_dum[df_dum['Id'] < 1461]
print(train_dum.shape)
display(train_dum.head())
```

```python
test_dum = df_dum[df_dum['Id'] > 1460]
test_dum = test_dum.drop('SalePrice', axis = 1)
print(test_dum.shape)
display(test_dum.head())
```

### Feature Scaling

```python
feature_scale = [feature for feature in df_dum.columns if feature not in ['Id', 'SalePrice']]
```

```python
scaler = MinMaxScaler()
X_train_dum = pd.DataFrame(scaler.fit_transform(train_dum[feature_scale]), columns = feature_scale)
display(X_train_dum.head())
print(X_train_dum.shape)
```

```python
X_test_dum = pd.DataFrame(scaler.transform(test_dum[feature_scale]), columns = feature_scale)
display(X_test_dum.head())
print(X_test_dum.shape)
```
